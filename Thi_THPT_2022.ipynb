# LÀM SẠCH VÀ TIỀN XỬ LÝ DỮU LIỆU

```python
import pandas as pd
file = "/content/drive/MyDrive/Colab Notebooks/Data Analytics/ThiTNTHPT 2022.csv"

df = pd.read_csv(file)

df.info()
```

```python
# Xem chiều dài của df, tương đương shape[0]
print('Len:', len(df))
# Xem kích thước của dataframe
print('Shape:', df.shape)
```

```python
# Xóa cột Tinh/Tp
df = df.drop('Tinh/Tp', axis=1)

df.info()
```

```python
# Đặt lại tên các cột
df = df.rename(columns = {'SBD':'SBD','Toan':'TOAN','Van':'VAN', 'Ngoai Ngu':'ANH','Ly':'LY', 'Hoa':'HOA', 'Sinh':'SINH',
                          'Su':'SU', 'Dia':'DIA', 'GDCD':'GDCD', 'Ma Tinh/TP':'MA TINH/TP'})
df
```

```python
# Thêm cột số thứ tự ở đầu DataFrame
df.insert(0, 'NO', range(1, len(df) + 1))

# Hiển thị DataFrame với cột số thứ tự
print(df)
```

```python
df.head(10)
```

```python
# Kiểm tra dữ liệu trùng lặp dự trên SBD
duplicate_rows = df[df.duplicated(subset="SBD")]
print('Số dòng dữ liệu trùng lặp: ', len(duplicate_rows))
```

```python
# Kiểm tra dữ liệu trống
df.isnull().sum()
```

```python
# Tạo DataFrame tổng hợp từ các DataFrame đã chuẩn hóa:
df_combined = pd.concat([df], ignore_index=True)
df_combined
```

```python
# Tạo DataFrame mới lưu trữ dòng dữ liệu không hợp lệ ban đầu
df_invalid = df_combined.copy()

# Hàm kiểm tra tính hợp lệ của cột RANK, NAME, COUNTRY
def is_df_data(row):
    if row['SBD'] == '0' or row['MA TINH/TP'] == '0':
        return False
    return True

# Áp dụng hàm kiểm tra tính hợp lệ cho từng dòng dữ liệu
mask = df.apply(lambda row: is_df_data(row), axis=1)

# Chia dữ liệu thành df_valid và df_invalid dựa trên kết quả kiểm tra
df_ThiTHPT2022 = df[mask]
df_ThiTHPT2022Loi = df[~mask]

df_ThiTHPT2022
```

```python
# Lưu DataFrame df_valid thành file Excel
df_ThiTHPT2022.to_csv('/content/drive/MyDrive/Colab Notebooks/Data Analytics/df_ThiTHPT2022.csv', index=False)

# Lưu DataFrame df_invalid thành file Excel
df_ThiTHPT2022Loi.to_csv('/content/drive/MyDrive/Colab Notebooks/Data Analytics/df_ThiTHPT2022Loi.csv', index=False)
```

**KHỐI A**

```python
import pandas as pd
import numpy as np

# Đọc dữ liệu từ file CSV vào DataFrame
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Analytics/df_ThiTHPT2022.csv')

#Lấy cột Toán, Lý và Hóa ra, tạo một dataframe mới
df1 = df[["SBD", "TOAN", "LY", "HOA"]]

# Xóa những thí sinh không thi khối A
df1 = df1.dropna(subset=['TOAN', 'LY', 'HOA'])

df1
```

```python
# Tính điểm thi khối A
df1['KHOI A'] = df1['TOAN'] + df1['LY'] + df1['HOA']

df1
```

```python
# Sắp xếp giảm dần theo cột 'KHOI A' để tìm thủ khoa
Thu_Khoa_A = df1.sort_values(by='KHOI A',
                             ascending=False).iloc[0]

# Hiển thị thông tin thủ khoa
print("Thông tin thủ khoa khối A:")
print(Thu_Khoa_A)
```

```python
#Lấy cột KHOI A ra, tạo một dataframe mới
df_Khoi_A = df1[["SBD", "KHOI A"]]
df_Khoi_A
```

```python
# Thống kê tần suất xuất hiện điểm
frequency_table = df_Khoi_A['KHOI A'].value_counts().reset_index()
frequency_table.columns = ['Giá trị điểm', 'Tần suất xuất hiện']
frequency_table = frequency_table.sort_values(by='Giá trị điểm')

df_Diem_Khoi_A = frequency_table
# Hiển thị bảng thống kê tần suất xuất hiện
print("Bảng thống kê tần suất xuất hiện của các giá trị điểm:")
df_Diem_Khoi_A
```

```python
import matplotlib.pyplot as plt
# Khoảng điểm
bins = 51

# Vẽ histogram từ DataFrame
plt.hist(df_Khoi_A['KHOI A'], bins=bins, edgecolor='black')
plt.xlabel('Điểm')
plt.ylabel('Số lượng thí sinh')
plt.title('Phổ điểm thi khối A năm 2022')
plt.show()
```

```python
df_Khoi_A['KHOI A'].describe().round(3)
```

```python
# Tính điểm trung bình, điểm trung vị, mode và độ lệch chuẩn
mean = np.mean(df_Khoi_A['KHOI A']).round(3)
median = np.median(df_Khoi_A['KHOI A'])
std = np.std(df_Khoi_A['KHOI A']).round(3)
mode = df_Khoi_A['KHOI A'].mode()

print("Điểm trung bình:", mean)
print("Điểm trung vị:", median)
print("Độ lệch chuẩn:", std)
print("Điểm xuất hiện nhiều nhất:", mode)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report
import pandas as pd

# Đọc dữ liệu từ tập tin excel
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data Analytics/df_ThiTHPT2022_Cluster.csv")

data
```

```python
# Xóa cột "NO" vì không cần thiết cho việc phân loại
data.drop("NO", axis=1, inplace=True)

data
```

```python
# Chia dữ liệu thành features (X) và nhãn (y)
X = data.drop("Cluster", axis=1)
y = data["Cluster"]
```

```python
# Chuyển đổi các cột dữ liệu dạng văn bản sang dạng số hóa
X = pd.get_dummies(X)

# Sử dụng KNN để phân loại
# Chia dữ liệu thành tập huấn luyện và tập kiểm tra (test data=train data)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Khởi tạo mô hình KNN với k=3
knn_model = KNeighborsClassifier(n_neighbors=2)
knn_model.fit(X_train, y_train)
```
