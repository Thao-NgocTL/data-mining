```python
df= df.rename(columns ={'Student ID':'SBD','Mathematics':'TOAN','Literature':'VAN','Foreign language':'ANH','Physics':'LY','Chemistry':'HOA','Biology':'SINH','History':'SU','Geography':'DIA','Civic education':'GDCD' })
df
```

```python
duplicated_rows = df[df.duplicated()]

# In các hàng trùng lặp (nếu có)
if not duplicated_rows.empty:
    print("Các hàng trùng lặp:")
    print(duplicated_rows)
else:
    print("Không có hàng trùng lặp trong dữ liệu.")
```

```python
maTinh = []
for i in df['SBD']:
  if len(str(i)) == 7:
    maTinh.append(str(i)[0])
  else:
    maTinh.append(str(i)[:2])

df['MaTinh'] = maTinh

df
```

```python
df.head(10)
```

```python
df.info()
```

```python
df["MaTinh"] = pd.to_numeric(df['MaTinh'],errors='coerce')
df.info()
```

```python
#Ktra dữ liệu null
df.isnull().sum()
```

```python
# Tạo DataFrame tổng hợp từ các DataFrame đã chuẩn hóa:
df_combined = pd.concat([df], ignore_index=True)
df_combined
```

```python
# Tạo DataFrame mới lưu trữ dòng dữ liệu không hợp lệ ban đầu
df_invalid = df_combined.copy()
# Hàm kiểm tra tính hợp lệ của cột RANK, NAME, COUNTRY
def is_df_data(row):
    if row['SBD'] == '0' or row['MaTinh'] == '0':
        return False
    return True
# Áp dụng hàm kiểm tra tính hợp lệ cho từng dòng dữ liệu
mask = df.apply(lambda row: is_df_data(row), axis=1)
# Chia dữ liệu thành df_valid và df_invalid dựa trên kết quả kiểm tra
df_ThiTHPT2023 = df[mask]
df_ThiTHPT2023Loi = df[~mask]

df_ThiTHPT2023
```

```python
# Lưu DataFrame df_valid thành file Excel
df_ThiTHPT2023.to_csv('/content/drive/MyDrive/Phân tích dữ liệu/ThiTHPT2023.csv', index=False)
# Lưu DataFrame df_invalid thành file Excel
df_ThiTHPT2023Loi.to_csv('/content/drive/MyDrive/Phân tích dữ liệu/ThiTHPT2023Loi.csv', index=False)
```

```python


import pandas as pd
import numpy as np
# Đọc dữ liệu từ file CSV vào DataFrame
df = pd.read_csv('/content/drive/MyDrive/Phân tích dữ liệu/ThiTHPT2023.csv')
#Lấy cột Toán, Lý và Hóa ra, tạo một dataframe mới
df1 = df[["SBD", "TOAN", "LY", "HOA"]]
# Xóa những thí sinh không thi khối A
df1 = df1.dropna(subset=['TOAN', 'LY', 'HOA'])
df1
```

```python
# Tính điểm thi khối A
df1['KHOI A'] = df1['TOAN'] + df1['LY'] + df1['HOA']
df1
```

```python
# Sắp xếp giảm dần theo cột 'KHOI A' để tìm thủ khoa
Thu_Khoa_A = df1.sort_values(by='KHOI A', ascending=False).iloc[0]
# Hiển thị thông tin thủ khoa
print("Thông tin thủ khoa khối A:")
print(Thu_Khoa_A)
```

```python
#Lấy cột KHOI A ra, tạo một dataframe mới
df_Khoi_A = df1[["SBD", "KHOI A"]]
df_Khoi_A
```

```python
# Thống kê tần suất xuất hiện điểm
frequency_table = df_Khoi_A['KHOI A'].value_counts().reset_index()
frequency_table.columns = ['Giá trị điểm', 'Tần suất xuất hiện']
frequency_table = frequency_table.sort_values(by='Giá trị điểm')
df_Diem_Khoi_A = frequency_table
# Hiển thị bảng thống kê tần suất xuất hiện
print("Bảng thống kê tần suất xuất hiện của các giá trị điểm:")
df_Diem_Khoi_A
```

```python
import matplotlib.pyplot as plt
# Khoảng điểm
bins = 51
# Vẽ histogram từ DataFrame
plt.hist(df_Khoi_A['KHOI A'], bins=bins, edgecolor='black')
plt.xlabel('Điểm')
plt.ylabel('Số lượng thí sinh')
plt.title('Phổ điểm thi khối A năm 2023')
plt.show()
```

```python
df_Khoi_A['KHOI A'].describe().round(3)
```

```python
# Tính điểm trung bình, điểm trung vị, mode và độ lệch chuẩn
mean = np.mean(df_Khoi_A['KHOI A']).round(3)
median = np.median(df_Khoi_A['KHOI A'])
std = np.std(df_Khoi_A['KHOI A']).round(3)
mode = df_Khoi_A['KHOI A'].mode()
print("Điểm trung bình:", mean)
print("Điểm trung vị:", median)
print("Độ lệch chuẩn:", std)
print("Điểm xuất hiện nhiều nhất:", mode)
```

```python
df_Toan['TOAN'].describe().round(3)
```

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Phân tích dữ liệu/ThiTHPT2023.csv")

df.info()

df
```

```python
# Tính điểm KHTN và KHXH
df['KHTN']=((df['HOA'] + df['LY'] + df['SINH'])/3).round(3)

df['KHXH']=((df['SU'] + df['DIA'] + df['GDCD'])/3).round(3)

my_df = df

my_df
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Tiền xử lý dữ liệu
my_df_khtn.fillna(my_df_khtn.mean(), inplace=True)

# Tìm số cụm tối ưu sử dụng phương pháp Elbow
max_clusters = 10  # Số cụm tối đa muốn xem
sse = []
for k in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(my_df_khtn)
    sse.append(kmeans.inertia_)
# Vẽ biểu đồ Elbow
plt.plot(range(1, max_clusters + 1), sse, marker='o')
plt.xlabel('Số lượng cụm')
plt.ylabel('SSE')
plt.title('Phương pháp Elbow')
plt.show()
```

```python
# Số cụm tối ưu dựa trên biểu đồ Elbow
num_clusters = 6

# Áp dụng thuật toán K-means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(my_df_khtn)

# Gán nhãn cụm cho dữ liệu
my_df_khtn['Cluster'] = kmeans.labels_

# Vẽ biểu đồ phân cụm
plt.scatter(my_df_khtn['Cluster'], my_df_khtn['TOAN'], c=my_df_khtn['Cluster'], cmap='viridis')
plt.xlabel('Cluster')
plt.ylabel('Toán')
plt.title('Phân cụm K-means')
plt.show()
```

```python

# Xóa cột "Unnamed: 0" vì không cần thiết cho việc phân loại
data.drop("Unnamed: 0", axis=1, inplace=True)

data
```

```python
# Chia dữ liệu thành features (X) và nhãn (y)
X = data.drop("Cluster", axis=1)
y = data["Cluster"]
```

```python
# Chuyển đổi các cột dữ liệu dạng văn bản sang dạng số hóa
X = pd.get_dummies(X)

# Sử dụng KNN để phân loại
# Chia dữ liệu thành tập huấn luyện và tập kiểm tra (test data=train data)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Khởi tạo mô hình KNN với k=4
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train, y_train)
```

```python
# Dự đoán nhãn cho tập kiểm tra
y_pred = knn_model.predict(X_test)

# Đánh giá mô hình KNN
print("KNN Model Evaluation:")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Xây dựng mô hình KNN với K lần lượt là 3, 5, 7, 9 trên tập train:
from sklearn.neighbors import KNeighborsClassifier

knn_3 = KNeighborsClassifier(n_neighbors=3)
knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_7 = KNeighborsClassifier(n_neighbors=7)
knn_9 = KNeighborsClassifier(n_neighbors=9)
knn_11 = KNeighborsClassifier(n_neighbors=11)

knn_3.fit(X_train, y_train)
knn_5.fit(X_train, y_train)
knn_7.fit(X_train, y_train)
knn_9.fit(X_train, y_train)
knn_11.fit(X_train, y_train)

# Đánh giá mô hình trên tập test:
y_pred_3 = knn_3.predict(X_test)
y_pred_5 = knn_5.predict(X_test)
y_pred_7 = knn_7.predict(X_test)
y_pred_9 = knn_9.predict(X_test)
y_pred_11 = knn_11.predict(X_test)

# Tính toán sai số trên tập train và test:
from sklearn.metrics import accuracy_score

train_error_3 = 1 - accuracy_score(y_train, knn_3.predict(X_train))
test_error_3 = 1 - accuracy_score(y_test, y_pred_3)

train_error_5 = 1 - accuracy_score(y_train, knn_5.predict(X_train))
test_error_5 = 1 - accuracy_score(y_test, y_pred_5)

train_error_7 = 1 - accuracy_score(y_train, knn_7.predict(X_train))
test_error_7 = 1 - accuracy_score(y_test, y_pred_7)

train_error_9 = 1 - accuracy_score(y_train, knn_9.predict(X_train))
test_error_9 = 1 - accuracy_score(y_test, y_pred_9)

train_error_11 = 1 - accuracy_score(y_train, knn_11.predict(X_train))
test_error_11 = 1 - accuracy_score(y_test, y_pred_11)

# Vẽ đồ thị sai số:
import matplotlib.pyplot as plt

plt.plot([3,5,7,9,11], [train_error_3, train_error_5, train_error_7, train_error_9, train_error_11], label='Train error')
plt.plot([3,5,7,9,11], [test_error_3, test_error_5, test_error_7, test_error_9, test_error_11], label='Test error')

plt.xlabel('K')
plt.ylabel('Error')
plt.legend()

plt.show()
```
